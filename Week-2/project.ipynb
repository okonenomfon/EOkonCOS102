```python
print("Welcome to the root's calculator!!!\nI solve quadratic, cubic and quartic equations only")
print("Which equation's root would you love to find?\n")

r = input("What is your choice?\n(Input the equation type in lowercase alphabet.)")

if r == 'quadratic':
    # ax^2 + bx + c = 0
    a = float(input("Enter coefficient a: "))
    b = float(input("Enter coefficient b: "))
    c = float(input("Enter coefficient c: "))

    d = b ** 2 - 4 * a * c
    if d > 0:
        r1 = (-b + d ** 0.5) / (2 * a)
        r2 = (-b - d ** 0.5) / (2 * a)
        print(f"The roots are {r1} and {r2}")
    elif d == 0:
        r1 = -b / (2 * a)
        print(f"The root is {r1} twice.")
    else:
        print("Complex roots")

elif r == 'cubic':
    # ax^3 + bx^2 + cx + d = 0
    a = float(input("Enter coefficient a: "))
    b = float(input("Enter coefficient b: "))
    c = float(input("Enter coefficient c: "))
    d = float(input("Enter coefficient d: "))

    p = (3 * a * c - b ** 2) / (3 * a ** 2)
    q = (2 * b ** 3 - 9 * a * b * c + 27 * a ** 2 * d) / (27 * a ** 3)
    x = (q ** 2) / 4 + (p ** 3) / 27

    if x > 0:
        r = (-q / 2 + x ** 0.5) ** (1 / 3)
        s = (-q / 2 - x ** 0.5) ** (1 / 3)
        real = -b / (3 * a)
        r1 = r + s + real
        print(f"The root is {r1}")
    elif x == 0:
        r = (-q / 2) ** (1 / 3)
        r1 = 2 * r - b / (3 * a)
        r2 = -r - b / (3 * a)
        print(f"The roots are {r1} and {r2}")
    else:
        r = (-q / 2 + x ** 0.5) ** (1 / 3)
        s = (-q / 2 - x ** 0.5) ** (1 / 3)
        real = -b / (3 * a)
        r1 = r + s + real
        print(f"The root is {r1}")

elif r == 'quartic':
    # ax^4 + bx^3 + cx^2 + dx + e = 0
    a = float(input("Enter coefficient a: "))
    b = float(input("Enter coefficient b: "))
    c = float(input("Enter coefficient c: "))
    d = float(input("Enter coefficient d: "))
    e = float(input("Enter coefficient e: "))

    p = (8 * a * c - 3 * b ** 2) / (8 * a ** 2)
    q = (b ** 3 - 4 * a * b * c + 8 * a ** 2 * d) / (8 * a ** 3)
    r = (16 * a * b * c - 64 * a ** 2 * d - 3 * b ** 4 + 256 * a ** 3 * e) / (256 * a ** 4)

    z = (p ** 2 / 4 + q + r) ** 0.5

    A = (-p / 2 + z) ** 0.5
    B = (-p / 2 - z) ** 0.5

    r1 = -b / (4 * a) - (A + B) / 2
    r2 = -b / (4 * a) - (A - B) / 2
    r3 = -b / (4 * a) - (-A + B) / 2
    r4 = -b / (4 * a) - (-A - B) / 2

    print(f"The roots are {r1}, {r2}, {r3}, and {r4}.")

else:
    print("Quadratic, Cubic and Quartic equations only.")
```
